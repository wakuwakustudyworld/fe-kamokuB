## 7-2-1　予想問題
### 問16　Min-Maxスケーリング
### 〔プログラム〕
<pre>
○整数型の配列: normalize(整数型の配列: Data, 整数型: n, 整数型の配列: Result)
　整数型: i, min_val, max_val, range_val
　
　min_val ← Data[1]
　max_val ← Data[1]
　
　for (i を 2 から n まで 1 ずつ増やす)
　　if (Data[i] < min_val)
　　　min_val ← Data[i]
　　endif
　　if (Data[i] > max_val)
　　　max_val ← Data[i]
　　endif
　endfor
　
　range_val ← <span style="border: solid 1px">　a　</span>
　
　for (i を 1 から n まで 1 ずつ増やす)
　　if (range_val = 0)
　　　Result[i] ← 50
　　else
　　　Result[i] ← <span style="border: solid 1px">　b　</span>
　　endif
　endfor
　
　return Result
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def normalize(Data: Array, n: int, Result: Array) -> Array:
    min_val = Data[1]
    max_val = Data[1]
    for i in range(2, n + 1):
        if Data[i] < min_val:
            min_val = Data[i]
        if Data[i] > max_val:
            max_val = Data[i]
    range_val = max_val - min_val
    for i in range(1, n + 1):
        if range_val == 0:
            Result[i] = 50
        else:
            Result[i] = (Data[i] - min_val) * 100 / range_val
    return Result
```

```python
# 実行例
Data = Array(5, 0) # サイズ5の配列を作成し、全ての要素を0で初期化
Data[1] = 10
Data[2] = 12
Data[3] = 23
Data[4] = 23
Data[5] = 16
n = 5
Result = Array(5, 0) # 結果を格納する配列
print(normalize(Data, n, Result))  # 出力: [0.0, 15.384615384615385, 100.0, 100.0, 46.15384615384615]
```
