## 7-2-1　予想問題
### 問7
### 〔プログラム〕
<pre>
○整数型の配列: reverseArray(整数型の配列: data, 整数型: n)
　整数型の配列: stack ← {n個の未定義の値}
　整数型: top ← 0
　整数型: i
　for (i を 1 から n まで 1 ずつ増やす)
　　top ← top + 1
　　stack[top] ← data[i]
　endfor
　for (i を 1 から n まで 1 ずつ増やす)
　　data[i] ← <span style="border: solid 1px">　a　</span>
　　top ← <span style="border: solid 1px">　b　</span>
　endfor
　return data
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def reverseArray(data: Array, n: int) -> Array:
    stack = Array(n, init=None)
    top = 0
    i = None
    for i in range(1, n + 1):
        top = top + 1
        stack[top] = data[i]
    for i in range(1, n + 1):
        data[i] = stack[top]
        top = top - 1
    return data
```

```python
# 実行例
data = Array.from_literal([10, 12, 12, 13, 12, 11, 12, 14, 13, 100])
n = len(data)
result = reverseArray(data, n)
print(result)
```
