## 3-2-3 初期値と終値

### 初期値と終値
【例】初期値が1，終値が3のfor文  

```python
for i in range(1, 4):
    print(i)
```

### 配列と初期値，終値

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array

array = Array.from_literal([1, 2, 3, 4, 5])
```

【例】配列arrayの要素をすべて出力（要素番号が1から始まる場合）  

```python
for i in range(1, len(array)+1):
    print(array[i])
```

配列の要素が0で始まる場合（Pythonの標準的な配列）

```python
array = [1, 2, 3, 4, 5]
```

同じプログラムを実行するとエラーとなります。

```python
for i in range(1, len(array)+1):
    print(array[i])
```

【例】要素番号が0で始まる場合のfor文の修正

```python
for i in range(0, len(array)):
    print(array[i])
```

### 配列の先頭から始まらない場合
3-2-1のトレース（基本情報技術者試験（科目B試験）サンプル問題セット 問3）を再掲

```python
def makeNewArray(ina : Array) -> Array:
    out = Array() # 要素数0の配列
    out.append(ina[1])
    for i in range(2, len(ina)+1):
        tail = out[len(out)]
        out.append(tail + ina[i])
    return out
```
