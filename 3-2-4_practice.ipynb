### 問1　関数fの実行結果
Pythonでは，余り（mod）は「%」で表します。

```python
def f(x : int, y : int) -> int:
    if y == 0:
        return x
    else:
        return f(y, x % y)
```

```python
f(775, 527)
```

### 問2 印字される数字の順番
〔プログラム〕

```python
def proc(n):
    if n == 0:
        return
    else:
        print(n, end = '') # 改行なしで出力
        proc(n - 1)
        print(n, end = '') # 改行なしで出力
        return
```

```python
proc(5)
```

### 問3 流れ図の初期値

```python
def f(a, b):
    print("① → ", end = '')
    m = a
    n = b
    while True:
        print("② → ", end = '')
        if m == n:
            break
        print("③ → ", end = '')
        if m > n:
            print("④ → ", end = '')
            m = m - n
        else:
            print("⑤ → ", end = '')
            n = n - m
    print("⑥")
    print(f"mの値 {m}")
```

```python
# 実行例 3a=2bとなるように，a=2, b=3とする。
f(2, 3)
```
