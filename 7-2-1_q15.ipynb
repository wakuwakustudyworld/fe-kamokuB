## 7-2-1　予想問題
### 問15
### 〔プログラム〕
<pre>
○整数型: detectOutliers(実数型の配列: data, 整数型: n, 実数型: k)
　実数型: mean, variance, threshold
　実数型: sum ← 0
　整数型: i, outlierCount
　for (iを1からnまで1ずつ増やす)
　　sum ← sum + data[i]
　endfor
　mean ← sum ÷ n
　
　variance ← 0
　for (iを1からnまで1ずつ増やす)
　　variance ← variance + <span style="border: solid 1px">　a　</span>
　endfor
　variance ← variance ÷ n
　
　threshold ← (k × k) × variance
　outlierCount ← 0
　for (iを1からnまで1ずつ増やす)
　　if (<span style="border: solid 1px">　b　</span>)
　　　outlierCount ← outlierCount + 1
　　endif
　endfor
　return outlierCount
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def detectOutliers(data: Array, n: int, k: float) -> int:
    sum = 0
    for i in range(1, n + 1):
        sum = sum + data[i]
    mean = sum / n
    variance = 0
    for i in range(1, n + 1):
        variance = variance + (data[i] - mean) * (data[i] - mean)
    variance = variance / n
    threshold = k * k * variance
    outlierCount = 0
    for i in range(1, n + 1):
        if (data[i] - mean) * (data[i] - mean) > threshold:
            outlierCount = outlierCount + 1
    return outlierCount
```

```python
# 実行例
data = Array.from_literal([10, 12, 12, 13, 12, 11, 12, 14, 13, 100])
n = len(data)
k = 2.0
print(detectOutliers(data, n, k))  # 出力: 1 (100が外れ値)
```
