4. ウォーカー  
事前に，クラスリストを定義し，インスタンスを作成します。

```python
class ListElement:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
# あらかじめ，エレメント1，2を作成しておきます。
element1 = ListElement(1)
element2 = ListElement(2)
# 先頭エレメントをrootにします。
root = element1
# エレメント1のnextにエレメント2をつなげます。
element1.next = element2
```

```python
# 教科書のプログラムを実行
current = root
while current is not None:
    print(current.value)
    current = current.next
```

6. 最も重要な値の保持者

```python
# あらかじめ，要素番号が1から始まる配列valuesを作成しておきます
from one_based_array import Array
values = Array.from_literal([0, 3, 2, 4, 1])

# 教科書のプログラムを実行
max_value = values[1] # 最初の値を最大値とする
for i in range(2, len(values) + 1):
    if values[i] > max_value:
        max_value = values[i]

# 最大値を出力
print(max_value)
```

9. フォロワー  
6と同じ変数valuesを使用します。

```python
previousValue = None
currentValue = values[1] # 最初の値をcurrentValueとする
for i in range(2, len(values) + 1):
    previousValue = currentValue
    currentValue = values[i]
    if currentValue > previousValue:
        print(currentValue)
```

10. オーガナイザー

```python
# 先に要素番坊が1から始まる配列dataを作成しておきます。
data = Array.from_literal([2, 6, 3, 1, 4, 5])

n = len(data)
# 要素数nの配列binsを作成し，すべての要素をNoneで初期化します。
bins = Array.from_literal([None] * n) 

for i in range(1, n + 1):
    bins[data[i]] = data[i]

# binsの要素を順番に出力します。
for i in range(1, n + 1):
    print(bins[i])
```

11. テンポラリ  
6と9と同じ変数valuesを使用します。

```python
for i in range(2, len(values) + 1):
    if values[i] > values[i - 1]:
        temp = values[i]
        values[i] = values[i - 1]
        values[i - 1] = temp
```
