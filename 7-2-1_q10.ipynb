## 7-2-1　予想問題
### 問10　二分探索木への要素挿入
## 〔プログラム〕
<pre>
○整数型の配列: insertTree(整数型の配列: Tree, 整数型: maxSize, 整数型: value)
　整数型: pos
　pos ← 1
　while (pos ≦ maxSize かつ Tree[pos] ≠ 0)
　　if (<span style="border: solid 1px">　　　</span>)
　　　pos ← pos × 2
　　else
　　　pos ← pos × 2 + 1
　　endif
　endwhile
　if (pos ≦ maxSize)
　　Tree[pos] ← value
　endif
　return Tree
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def insertTree(Tree : Array, maxSize: int, value: int) -> Array:
    pos = 1
    while pos <= maxSize and Tree[pos] != 0:
        if value < Tree[pos]:
            pos = pos * 2
        else:
            pos = pos * 2 + 1
    if pos <= maxSize:
        Tree[pos] = value
    return Tree
```

```python
# 実行例
Tree = Array(15, 0) # サイズ15の配列を作成し、全ての要素を0で初期化
Tree = insertTree(Tree, 15, 10)
Tree = insertTree(Tree, 15, 5)
Tree = insertTree(Tree, 15, 15)
print(Tree)
```
