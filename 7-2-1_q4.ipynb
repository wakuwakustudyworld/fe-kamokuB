## 7-2-1　予想問題
### 問4　最小公倍数を求める関数
### 〔プログラム〕
<pre>
○整数型: gcd(整数型: a, 整数型: b)
　整数型: temp
　while (b ≠ 0)
　　temp ← b
　　b ← a mod b
　　a ← temp
　endwhile
　return a

○整数型: lcm(整数型: a, 整数型: b)
　整数型: g
　g ← gcd(a, b)
　return <span style="border: solid 1px">　　　</span>
</pre>


```python
def gcd(a: int, b: int) -> int:
    while b != 0:
        temp = b
        b = a % b
        a = temp
    return a

def lcm(a: int, b: int) -> int:
    g = gcd(a, b)
    return (a // g) * b
```

```python
# 実行例
print(gcd(48, 18))  # 出力: 6
print(lcm(48, 18))  # 出力: 144
```
