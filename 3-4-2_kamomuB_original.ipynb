### 問1 最も価値の高い要素の要素番号を返す関数
〔プログラム〕 
<pre>
○整数型: findMaxValueIndex(整数型の配列: values, 整数型: n)
  整数型: i, maxValue, maxIndex, currentValue
  maxValue ← values[1] × 1
  maxIndex ← 1
  
  for (i を 2 から n まで 1 ずつ増やす)
    currentValue ← values[i] × i
    if (currentValue > maxValue)
      maxValue ← currentValue
      maxIndex ← i
    endif
  endfor
  
  return maxIndex
</pre>


```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def findMaxValueIndex(values: Array, n: int) -> int:
    maxValue = values[1] * 1
    maxIndex = 1
    for i in range(2, n + 1):
        currentValue = values[i] * i # ここの右辺が空欄
        if currentValue > maxValue:
            maxValue = currentValue
            maxIndex = i
    return maxIndex
```

```python
# 実行例
values = Array.from_literal([1, 3, 2, 5, 4])
n = len(values)
result = findMaxValueIndex(values, n)
print(result)  # 出力: 4
```

### 問5 プログラムのデバッグ
〔プログラム〕（修正前）
<pre>
○整数型: findValue(整数型の配列: Data, 整数型: n, 整数型: target)
  整数型: i
  i ← 1
  
  while (i < n)
    if (Data[i] が target と 等しい)
      return i
    endif
    i ← i + 1
  endwhile
  
  return 0
</pre>

```python
def findValue(Data: Array, n: int, target: int) -> int:
    i : int = 1
    while i < n:
        if Data[i] == target:
            return i
        i = i + 1
    return 0
```

```python
# 実行例
values = Array.from_literal([1, 2, 3])
n = len(values)
target = 3 # 最後の要素を検索

result = findValue(values, n, target)
print(result)  # 出力: 3が正しいはず
```

修正後のプログラム

```python
def findValue(Data: Array, n: int, target: int) -> int:
    i : int = 1
    while i <= n:
        if Data[i] == target:
            return i
        i = i + 1
    return 0
```

```python
# 再実行
result = findValue(values, n, target)
print(result)  # 出力: 3が正しいはず
```
