## 7-2-1　予想問題
### 問1
### 〔プログラム〕
<pre>
○整数型: processValues(整数型: value1, 整数型: value2, 整数型: value3)
　整数型: max, min, count, result
　max ← value1
　min ← value1
　count ← 0
　if (value1 ≧ 10)
　　count ← count + 1
　endif
　if (value2 > max)
　　max ← value2
　endif
　if (value2 < min)
　　min ← value2
　endif
　if (value2 ≧ 10)
　　count ← count + 1
　endif
　if (value3 > max)
　　max ← value3
　endif
　if (value3 < min)
　　min ← value3
　endif
　if (value3 ≧ 10)
　　count ← count + 1
　endif
　result ← max - min
　if (count ≧ 2)
　　result ← <span style="border: solid 1px">　　　</span>
　endif
　return result
</pre>

```python
def processValues(value1, value2, value3):
    max = value1
    min = value1
    count = 0
    if value1 >= 10:
        count = count + 1
    if value2 > max:
        max = value2
    if value2 < min:
        min = value2
    if value2 >= 10:
        count = count + 1
    if value3 > max:
        max = value3
    if value3 < min:
        min = value3
    if value3 >= 10:
        count = count + 1
    result = max - min
    if count >= 2:
        result = result + 10
    return result
```

```python
# 実行例
print(processValues(12, 8, 15))  # 出力: 17
print(processValues(5, 3, 7))     # 出力: 4
```
