## 7-2-1　予想問題
### 問6 フィボナッチ数列を再帰的に求める関数
### 〔プログラム〕
<pre>
○整数型: fibonacci(整数型: n)
　if (n ≦ <span style="border: solid 1px">　a　</span>)
　　return 1
　endif
　return <span style="border: solid 1px">　b　</span>
</pre>


```python
def fibonacci(n):
    if n <= 2:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
# 実行例
print(fibonacci(5))  # 出力: 5
print(fibonacci(10)) # 出力: 55
```
