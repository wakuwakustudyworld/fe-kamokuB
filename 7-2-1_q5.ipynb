## 7-2-1　予想問題
### 問5
<pre>
○整数型: maxSubarraySum(整数型の配列: Data, 整数型: n)
　整数型: currentSum, maxSum, i
　currentSum ← Data[1]
　maxSum ← Data[1]
　
　for (i を 2 から n まで 1 ずつ増やす)
　　if (<span style="border: solid 1px">　a　</span>)
　　　currentSum ← Data[i]
　　else
　　　currentSum ← currentSum + Data[i]
　　endif
　　if (<span style="border: solid 1px">　b　</span>)
　　　maxSum ← currentSum
　　endif
　endfor
　
　return maxSum
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def maxSubarraySum(Data: Array, n: int) -> int:
    currentSum = Data[1]
    maxSum = Data[1]
    for i in range(2, n + 1):
        if currentSum < 0:
            currentSum = Data[i]
        else:
            currentSum = currentSum + Data[i]
        if currentSum > maxSum:
            maxSum = currentSum
    return maxSum
```

```python
Data = Array.from_literal([-2, 1, -3, 4, -1, 2, 1, -5, 4])
n = len(Data)
print(maxSubarraySum(Data, n))
```
