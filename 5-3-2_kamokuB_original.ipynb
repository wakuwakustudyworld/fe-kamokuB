## 問1 移動平均を求める関数

<pre>
○実数型の配列: movingAverage(整数型の配列: data, 整数型: n, 整数型: windowSize)
  実数型の配列: result ← {} 
  整数型: i, j
  実数型: sum
  
  for (i を 1 から n - windowSize + 1 まで 1 ずつ増やす)
    sum ← 0
    for (j を i から i + windowSize - 1 まで 1 ずつ増やす)
      sum ← sum + data[j]
    endfor
    resultの末尾 に sum ÷ windowSize を追加する
  endfor
  
  return result
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def movingAverage(data: Array, n: int, windowSize: int) -> Array:
    result = Array.from_literal([])
    for i in range(1, n - windowSize + 1 + 1):
        sum = 0
        for j in range(i, i + windowSize - 1 + 1):
            sum = sum + data[j]
        result.append(sum / windowSize)
    return result
```

```python
data = Array.from_literal([1, 2, 3, 4, 5])
n = 5
windowSize = 3
print(movingAverage(data, n, windowSize))
```
