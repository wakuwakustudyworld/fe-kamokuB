## 3-2-2 既存のプログラムの変更
### 最大値プログラムを最小値プログラムに変更


```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

〔最大値を求めるプログラム〕（再掲：トレースのprint文なし）

```python
def findMax(array : Array)  -> int:
    maxPos : int = 1
    for i in range(2, len(array) + 1):
        if array[i] > array[maxPos]:
            maxPos = i
    return maxPos
```

〔最小値を求めるプログラム1〕

```python
def findMax(array : Array)  -> int:
    maxPos : int = 1
    for i in range(2, len(array) + 1):
        if array[i] < array[maxPos]:
            maxPos = i
    return maxPos
```

〔最小値を求めるプログラム2〕

```python
def findMin(array : Array)  -> int:
    minPos : int = 1
    for i in range(2, len(array) + 1):
        if array[i] < array[minPos]:
            minPos = i
    return minPos
```

例の引数で試してみる

```python
array = Array.from_literal([0.4, 0.4, 4.2, 1.8, 3.5])

findMin(array)
```

より小さいを以下に変更した例

```python
def findMin(array : Array)  -> int:
    minPos : int = 1
    for i in range(2, len(array) + 1):
        if array[i] <= array[minPos]:
            minPos = i
    return minPos
```

変更した関数を実行

```python
findMin(array)
```
