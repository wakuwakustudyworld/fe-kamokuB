## 7-2-1　予想問題
### 問14　分散の計算と散らばり度の判定
### 〔プログラム〕
<pre>
○整数型: evaluateVariance(整数型の配列: Data, 整数型: n)
　整数型: i
　整数型: sum ← 0
　実数型: mean
　実数型: variance ← 0.0
　実数型: deviation
　for (i を 1 から n まで 1 ずつ増やす)
　　sum ← sum + Data[i]
　endfor
　mean ← sum ÷ n
　for (i を 1 から n まで 1 ずつ増やす)
　　deviation ← Data[i] - mean
　　variance ← variance + <span style="border: solid 1px">　a　</span>
　endfor
　variance ← variance ÷ n
　if (variance < 10)
　　return 1
　elseif (<span style="border: solid 1px">　b　</span>)
　　return 2
　else
　　return 3
　endif
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def evaluateVariance(Data: Array, n: int) -> int:
    i = None
    sum = 0
    mean = None
    variance = 0.0
    deviation = None
    for i in range(1, n + 1):
        sum = sum + Data[i]
    mean = sum / n
    for i in range(1, n + 1):
        deviation = Data[i] - mean
        variance = variance + deviation * deviation
    variance = variance / n
    if variance < 10:
        return 1
    elif variance < 50:
        return 2
    else:
        return 3
```

```python
## 実行例
Data = Array(5, 0) # サイズ5の配列を作成し、全ての要素を0で初期化
Data[1] = 10
Data[2] = 12
Data[3] = 23
Data[4] = 23
Data[5] = 16
n = 5
print(evaluateVariance(Data, n))  # 出力: 2
```
