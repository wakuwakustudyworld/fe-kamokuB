## 3-1-1 既存のプログラムを読むこと  
### プログラムを読むために大切なこと
トレース用にprint文を追加

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def findMax(array : Array)  -> int:
    maxPos : int = 1
    for i in range(2, len(array) + 1):
        # トレース用に追加
        print(f"i={i}, MaxPos={maxPos}, array[i]={array[i]}, array[maxPos]={array[maxPos]}, 判定: {array[i] > array[maxPos]}")
        # トレース用に追加　ここまで
        if array[i] > array[maxPos]:
            maxPos = i
    return maxPos
```

### トレース
例の引数で実行

```python
array = Array.from_literal([3.5, 2.5, 4.2, 1.8, 0.4])

print(f'最終結果: {findMax(array)}')
```
