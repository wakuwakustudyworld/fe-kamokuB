### 問3
### 〔プログラム〕
<pre>
○整数型の配列: rotateLeft(整数型の配列: Data, 整数型: n)
　整数型: i, temp
　<span style="border: solid 1px">　　　</span>
　for (i を 1 から n - 1 まで 1 ずつ増やす)
　　Data[i] ← Data[i + 1]
　endfor
　Data[n] ← temp
　return Data
</pre>

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array
```

```python
def rotateLeft(Data: Array, n: int) -> Array:
    temp = Data[1]
    for i in range(1, n - 1 + 1):
        Data[i] = Data[i + 1]
    Data[n] = temp
    return Data
```

```python
Data = Array.from_literal([5, 3, 8, 1])
n = len(Data)
print(rotateLeft(Data, n))
```
