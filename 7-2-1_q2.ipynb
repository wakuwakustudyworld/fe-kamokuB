○整数型: calculateElectricityBill(整数型: usage)
　整数型: total ← 0
　整数型: remaining ← usage
　if (remaining > 120)
　　total ← total + 120 × 20
　　remaining ← remaining - 120
　else
　　total ← total + remaining × 20
　　return total
　endif
　if (remaining > 180)
　　<span style="border: solid 1px">　　　</span>
　　remaining ← remaining - 180

　endif
　total ← total + remaining × 80

```python
def calculateElectricityBill(usage: int) -> int:
    total = 0
    remaining = usage
    if remaining > 120:
        total = total + 120 * 20
        remaining = remaining - 120
    else:
        total = total + remaining * 20
        return total
    if remaining > 180:
        total = total + 180 * 25
        remaining = remaining - 180
    else:
        total = total + remaining * 25
        return total
    total = total + remaining * 80
    return total
```

```python
calculateElectricityBill(310) # 120*20 + 180*25 + 10*80 = 2400 + 4500 + 800 = 7700
```
