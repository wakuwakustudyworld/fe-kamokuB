### 問1　文字列の連結プログラム
初期値設定

```python
# 1で始まる配列　（one_based_array.pyで定義）
from one_based_array import Array

# X，Yの値の例
X = Array.from_literal([1, 2, 3])
Y = Array.from_literal([4, 5, 6])

# Z 初期値0を設定した長さX+Yのリストを作成
Z = Array.from_literal([0] * (len(X) + len(Y)))
```

〔プログラム〕

```python
m = len(X)
n = len(Y)
for k in range(1, m + 1):
    Z[k] = X[k]      # 空欄a
for k in range(1, n + 1):
    Z[m + k] = Y[k]  # 空欄b
```

```python
print(Z)
```

### 問2 手続を実行した結果
初期値設定

```python
x = 7
y = 3
```

〔プログラム〕

```python
q = 0
r = x
while r >= y:
    r = r - y
    q = q + 1
```

```python
print(q, r)
```
