## 7-2-1　予想問題
### 問12
### 〔プログラム〕
<pre>
○論理型: isPerfectNumber(整数型: n)
　整数型: i, sum
　sum ← 0
　for (i を 1 から n - 1 まで 1 ずつ増やす)
　　if (<span style="border: solid 1px">　a　</span>)
　　　sum ← sum + i
　　endif
　endfor
　if (<span style="border: solid 1px">　b　</span>)
　　return true
　else
　　return false
　endif
</pre>


```python
def isPerfectNumber(n):
    sum = 0
    for i in range(1, n - 1 + 1):
        if n % i == 0:
            sum = sum + i
    if sum == n:
        return True
    else:
        return False
```

```python
print(isPerfectNumber(6))  # True
print(isPerfectNumber(28)) # True
print(isPerfectNumber(12)) # False
```
