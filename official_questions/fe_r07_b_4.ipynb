### 問4 線形探索を行う関数
次の記述中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。ここで，
配列の要素番号は1から始まる。

関数searchは，二つの文字型の配列を，それぞれ引数data及びkeyで受け取り，dataから，keyの要素の並びと同じ並びを全て探し，その先頭の要素番号を全て格納した配列を返す。見つからなければ，要素数 0の配列を返す。  
関数searchをsearch({"a", "b", "a", "b", "c", "a", "b", "c"}, {"a", "b", "c"})として呼び出すと， /*** β ***/ の行の条件式が真となる回数は <span style="border: solid 1px">　　　</span> 回である。

### 〔プログラム〕
<pre> 
/* keyは，要素数1以上の配列である */ 
○整数型の配列 : search(文字型の配列 : data,文字型の配列: key)
　整数型: i, j, lenData, lenKey
　整数型の配列: result ← {} //要素数0の配列 

　lenData ← dataの要素数 
　lenKey ← keyの要素数 
　/* (lenData ー lenKey ＋ 1) が0以下のときは繰返し処理を実行しない */ 
　for (i を 1 から (lenData ー lenKey　＋ 1) まで 1 ずつ増やす) 
　　for (jを 1 から lenKey まで 1ずつ増やす) // α 
　　　if (data[i ＋ j ー 1] が key[j]と等しい) /*** β ***/
　　　　if (jが lenKeyと等しい) 
　　　　　resultの末尾に iの値を追加する 
　　　　endif 
　　　else
　　　　αの行から始まる繰返し処理を終了する 
　　　endif 
　　endfor 
　endfor 
　return result
</pre>

### 解答群
ア 1　　イ 2　　ウ 3　　エ 4　　オ 5  
カ 6　　キ 7　　ク 8　　ケ 9　　コ 10 


```python
# 配列を1から始めるためのクラス
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def search(data : Array, key: Array):
    beta = 0 # βの比較回数をカウントする変数
    result = Array.from_literal([])
    lenData = len(data)
    lenKey = len(key)
    # (lenData ー lenKey ＋ 1) が0以下のときは繰返し処理を実行しない
    for i in range(1, lenData - lenKey + 1 + 1):
        for j in range(1, lenKey + 1):
            if data[i + j - 1] == key[j]:
                beta += 1 # if文が真になる回数（β）をカウント
                if j == lenKey:
                    result.append(i)
            else:
                break

    print("/*** β ***/ の行の条件式が真となる回数: ", beta)
    return result
```

```python
search(Array.from_literal(["a", "b", "a", "b", "c", "a", "b", "c"]), 
       Array.from_literal(["a", "b", "c"]))
```
