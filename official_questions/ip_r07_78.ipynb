### 問78
手続 calcMod3 を呼び出したときの出力はどれか。

〔プログラム〕
```text
○calcMod3()
 整数型: totalValue, i
 totalValue ← 0
 for (iを1から7まで1ずつ増やす)
   if (i÷3の余りが0と等しい)
     totalValue ← totalValue + i
   endif
 endfor
 totalValue を出力する
```

ア 2   イ 3   ウ 7   エ 9 

````python
def calcMod3():
    totalValue = None
    i = None
    totalValue = 0
    for i in range(1, 7 + 1):
        if i % 3 == 0:
            totalValue = totalValue + i
    print(totalValue)
````

````python
calcMod3()
````
