## 基本情報技術者試験（科目B試験）サンプル問題セット
### 問4
　次のプログラム中の  <span style="border: solid 1px">　a　</span> ～ <span style="border: solid 1px">　c　</span>  に入れる正しい答えの組合せを，解答群の中から選べ。

　関数gcd は，引数で与えられた二つの正の整数num1とnum2の最大公約数を，次の(1)～(3) の性質を利用して求める。  
(1) num1とnum2が等しいとき，num1とnum2の最大公約数はnum1である。  
(2) num1がnum2より大きいとき，num1とnum2の最大公約数は，(num1－num2)とnum2の最大公約数と等しい。  
(3) num2がnum1より大きいとき，num1とnum2の最大公約数は，(num2－num1)とnum1の最大公約数と等しい。

### 〔プログラム〕
<pre>
○整数型: gcd(整数型: num1, 整数型: num2)
　整数型: x ← num1
　整数型: y ← num2
　<span style="border: solid 1px">　a　</span> 
　if (<span style="border: solid 1px">　b　</span>)
　  x ← x － y
　else
　  y ← y － x
　endif
　<span style="border: solid 1px">　a　</span> 
　return x
</pre>
### 解答群
||a|b|c|
|---|---|---|---|
|ア|if (x ≠ y)|x ＜ y|endif|
|イ|if (x ≠ y)|x ＞ y|endif|
|ウ|while (x ≠ y)|x ＜ y|endwhile|
|エ|while (x ≠ y)|x ＞ y|endwhile|

※Pythonでは，空欄c部分(endifやendwhile)は不要。

```python
def gcd(num1 : int, num2 : int) -> int:
    x : int = num1
    y : int = num2
    while x != y: # この行が空欄a
        if x > y: # この行の条件式が空欄b
            x = x - y
        else:
            y = y - x
    return x
```

選択肢  
空欄aは ア，イが 「if x != y」，ウ，エが「while x != y」  
空欄bは ア，ウが 「x > y」，イ，エが「x < y」

```python
#例
print(gcd(12, 18)) # 6
```
