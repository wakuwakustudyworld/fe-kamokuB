### 問11
次の記述中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。ここで，配列の要素番号は1から始まる。

関数 binSortを binSort(<span style="border: solid 1px">　　　</span>)として呼び出すと，戻り値の配列には未定
義の要素は含まれておらず，値は昇順に並んでいる。

### 〔プログラム〕
<pre>
○整数型の配列 : binSort(整数型の配列 : data)整数型:n ← 　dataの要素数整数型の配列: bins ← {n個の未定義の値}
　整数型: i 
　for (iを 1 から n まで 1 ずつ増やす)
　  bins[data[i]] ← data[i]
　endfor 
  return bins
</pre>
### 解答群
ア {2, 6, 3, 1, 4, 5}  
イ {3, 1, 4, 4, 5, 2}  
ウ {4, 2, 1, 5, 6, 2}  
エ {5, 3, 4, 3, 2, 6} 

```python
# 配列の要素番号を1から始めるためのクラス
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def binSort(data : Array) -> Array:
    n = len(data)
    bins = Array(n)
    for i in range(1, n+1):
        bins[data[i]] = data[i]
    return bins
```

テスト

```python
# 解答群ア
binSort(Array.from_literal([2, 6, 3, 1, 4, 5]))
```

```python
# 解答群イ
binSort(Array.from_literal([3, 1, 4, 4, 5, 2]))
```

```python
# 解答群ウ
binSort(Array.from_literal([4, 2, 1, 5, 6, 2]))
```

```python
# 解答群エ
binSort(Array.from_literal([5, 3, 4, 3, 2, 6]))
```
