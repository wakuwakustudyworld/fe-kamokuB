### 問3
次の記述中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。ここで，配列の要素番号は1から始まる。

関数makeNewArrayは，要素数2以上の整数型の配列を引数にとり，整数型の配列を返す関数である。関数makeNewArrayを makeNewArray({3, 2, 1, 6, 5, 4})として呼び出したとき，戻り値の配列の要素番号5の値は <span style="border: solid 1px">　　　</span> となる。

### 〔プログラム〕
<pre>
○整数型の配列: makeNewArray(整数型の配列: in)
　整数型の配列: out ← {} //要素数0の配列
　整数型: i, tail 
　outの末尾 に in[1]の値 を追加する 
　for (i を 2 から inの要素数 まで 1 ずつ増やす) 
　  tail ← out[outの要素数] 
　  outの末尾に (tail ＋ in[i]) の結果を追加する 
　endfor
　return out
</pre>

### 解答群
ア 5  
イ 6  
ウ 9  
エ 11  
オ 12  
カ 17  
キ 21 

※inが予約語なので，変数名をinからinaに変更。

```python
# 配列を1-basedで扱うためのArrayクラスをインポート
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def makeNewArray(ina : Array) -> Array:
    out = Array() # 要素数0の配列
    out.append(ina[1])
    for i in range(2, len(ina)+1):
        tail = out[len(out)]
        out.append(tail + ina[i])
    return out
```

```python
out = makeNewArray(Array.from_literal([3, 2, 1, 6, 5, 4]))
print(out[5])
```
