### 問4 マージソートで整列する関数

関数mergeは，昇順に整列された整数型の配列 data1及び data2を受け取り，これらを併合してできる昇順に整列された整数型の配列を返す。関数 mergeを merge({2, 3}, {1, 4})として呼び出すと，/*** α ***/の行は<span style="border: solid 1px">　　　</span>。

### 〔プログラム〕
<pre>
○整数型の配列 : merge(整数型の配列 : data1,整数型の配列 : data2)
 整数型: n1 ← data1の要素数
 整数型: n2 ← data2の要素数
 整数型の配列: work ← {(n1＋n2)個の未定義の値}

### 解答群
ア 実行されない  
イ 1回実行される  
ウ 2回実行される  
エ 3回実行される

```python
# 配列を1から始めるためのクラス
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def merge(data1 : Array, data2: Array) -> Array:
    alpha = 0 # αの比較回数をカウントする変数
    n1 = len(data1)
    n2 = len(data2)
    work = Array((n1 + n2), init=None)
    i = 1
    j = 1
    k = 1
    while i <= n1 and j <= n2:
        if data1[i] <= data2[j]:
            work[k] = data1[i]
            i = i + 1
        else:
            work[k] = data2[j]
            j = j + 1
        k = k + 1
    while i <= n1:
        work[k] = data1[i]
        i = i + 1
        k = k + 1
    while j <= n2:
        alpha = alpha + 1 # 実行回数（α）をカウント
        work[k] = data2[j]
        j = j + 1
        k = k + 1

    print("αの行は", alpha, "回実行される")
    return work
```

```python
merge(Array.from_literal([2, 3]), Array.from_literal([1, 4]))
```
