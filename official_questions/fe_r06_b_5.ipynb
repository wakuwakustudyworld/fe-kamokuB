### 問5 関連度を求める手続

一度の注文で購入された商品のリストを，注文ごとに記録した注文データがある。表に，注文データの例を示す。

注文データから，商品xと商品yとが同一の注文で購入されやすい傾向を示す関連度$L_{xy}$を，次の式で計算する。  

$$L_{xy}=\frac{M_{xy} × 全注文数}{K_x × K_y}$$

ここで，$M_{xy}$ は商品x と商品y とが同一の注文で購入された注文数，$K_x$ は商品xが購入された注文数，$K_y$ は商品yが購入された注文数を表す。表の例では，$M_{AB}$ が2，全注文数が6，$K_A$ が4，$K_B$ が3であるので，商品Aと商品Bの関連度 $L_{AB}$ は，(2 × 6) ／ (4 × 3) ＝ 1.0 である。  
手続putRelatedItemは，大域変数ordersに格納された注文データを基に，引数で与えられた商品との関連度が最も大きい商品のうちの一つと，その関連度を出力する。プログラムでは，商品は文字列で表し，注文は購入された商品の配列，注文データは注文の配列で表している。注文データには2種類以上の商品が含まれるものとする。また，注文データにある商品以外の商品が，引数として与えられることはないものとする。

### 解答群
<table><tr><th></th><th>a</th><th>b</th><th>c</th></tr>
<tr><td>ア</td><td>arrayK[i]</td><td>arrayM[i]</td><td>allItemsの要素数</td></tr>
<tr><td>イ</td><td>arrayK[i]</td><td>arrayM[i]</td><td>ordersの要素数</td></tr>
<tr><td>ウ</td><td>arrayK[i]</td><td>arrayM[i]</td><td>otherItemsの要素数</td></tr>
<tr><td>エ</td><td>arrayM[i]</td><td>arrayK[i]</td><td>allItemsの要素数</td></tr>
<tr><td>オ</td><td>arrayM[i]</td><td>arrayK[i]</td><td>ordersの要素数</td></tr>
<tr><td>カ</td><td>arrayM[i]</td><td>arrayK[i]</td><td>otherItemsの要素数</td></tr></table>

```python
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
orders = Array.from_literal([
    Array.from_literal(["A", "B", "D"]), 
    Array.from_literal(["A", "D"]), 
    Array.from_literal(["A"]), 
    Array.from_literal(["A", "B", "E"]), 
    Array.from_literal(["B"]), 
    Array.from_literal(["C", "E"])])

def putRelatedItem(item: str) -> None:
    global orders # ordersはグローバル変数であることを宣言
    # ordersに含まれる文字列を重複なく辞書順に格納した配列
    allItems = Array.from_literal(sorted(set(x for order in orders for x in order)))
    # allItemsの複製から値がitemである要素を除いた配列
    otherItems = Array.from_literal([x for x in allItems if x != item])
    itemCount = 0
    arrayK = Array(len(otherItems), init=0)
    arrayM = Array(len(otherItems), init=0)
    maxL = float('-inf')

    for order in orders:
        # orderのいずれかの要素の値がitemの値と等しい
        if item in order:
            itemCount += 1
        for i in range(1, len(otherItems) + 1):
            # orderのいずれかの要素の値がotherItems[i]の値と等しい
            if otherItems[i] in order:
                if item in order:
                    arrayM[i] += 1
                arrayK[i] += 1
    for i in range(1, len(otherItems) + 1):
        valueL = arrayM[i] * len(orders) / (itemCount * arrayK[i])
        if valueL > maxL:
            maxL = valueL
            relatedItem = otherItems[i]
    print(relatedItem, maxL, sep=",")
```

```python
# 実行例
putRelatedItem("A")
```
