### 問62　配列を変換する関数
関数 convert は，整数型の配列を一定のルールで文字列に変換するプログラムである。関数 convert を convert(arrayInput) として呼び出したときの戻り値が “AABAB” になる引数 arrayInput の値はどれか。ここで，arrayInput の要素数は1以上とし，配列の要素番号は1から始まる。

〔プログラム〕
```text
○文字列型:convert(整数型の配列: arrayInput)
文字列型:stringOutput ← ""  // 空文字列を格納
整数型:i
for (iを1から arrayInput の要素数まで1ずつ増やす)
  if (arrayInput[i] が1と等しい)
    stringOutput の末尾に "A" を追加する
  else
    stringOutput の末尾に "B" を追加する
  endif
endfor
return stringOutput
```

ア {0, 0, 1, 2, 1}  
イ {0, 1, 2, 1, 1}  
ウ {1, 0, 1, 2, 0}  
エ {1, 1, 2, 1, 0}

````python
# 配列の要素を1から始めるための拡張
import sys
sys.path.append('..')

from one_based_array import Array
````

````python
def convert(arrayInput):
    stringOutput = ''
    i = None
    for i in range(1, len(arrayInput) + 1):
        if arrayInput[i] == 1:
            stringOutput += 'A'  # 文字列の末尾に追加
        else:
            stringOutput += 'B'  # 文字列の末尾に追加
    return stringOutput
````

````python
# 空欄ア {0, 0, 1, 2, 1} を配列に変換して関数に渡す
convert(Array.from_literal([0, 0, 1, 2, 1]))   
````

````python
# 空欄イ {0,1,2,1,1} を配列に変換して関数に渡す
convert(Array.from_literal([0, 1, 2, 1, 1]))
````

````python
# 空欄ウ　{1, 0, 1, 2, 0} を配列に変換して関数に渡す
convert(Array.from_literal([1, 0, 1, 2, 0]))
````

````python
# 空欄エ　{1, 1, 2, 1, 0} を配列に変換して関数に渡す
convert(Array.from_literal([1, 1, 2, 1, 0]))
````
