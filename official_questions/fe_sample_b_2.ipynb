### 問2
次のプログラム中の <span style="border: solid 1px">　a　</span> と <span style="border: solid 1px">　b　</span> に入れる正しい答えの組合せを，解答群の中から選べ。ここで，配列の要素番号は1から始まる。

次のプログラムは，整数型の配列arrayの要素の並びを逆順にする。

〔プログラム〕
<pre>
　整数型の配列: array ← {1, 2, 3, 4, 5}
　整数型: right, left
　整数型: tmp
　for (left を 1 から (arrayの要素数 ÷ 2 の商) まで 1 ずつ増やす)
　  right ← <span style="border: solid 1px">　a　</span>
　  tmp ← array[right]
　  array[right] ← array[left]
　  <span style="border: solid 1px">　b　</span> ← tmp
　endfor
</pre>
### 解答群
||a|b|
|---|---|---|
|ア|arrayの要素数 － left|array[left]|
|イ|arrayの要素数 － left|array[right]|
|ウ|arrayの要素数 － left ＋ 1|array[left]|
|エ|arrayの要素数 － left ＋ 1|array[right]|


```python
# 配列を1-basedで扱うためのArrayクラスをインポート
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
array : Array = Array.from_literal([1, 2, 3, 4, 5])
for left in range(1, len(array)//2 + 1):
    right = len(array) - left + 1 # 右辺が空欄a
    tmp = array[right]
    array[right] = array[left]
    array[left] = tmp # 左辺が空欄b
```
