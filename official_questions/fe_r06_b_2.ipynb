### 問2 2進数を求める関数
次のプログラム中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。  

関数convDecimalは，引数として与えられた，“0”と“1”だけから成る，1文字以上の文字列を，符号なしの2進数と解釈したときの整数値を返す。例えば，引数として“10010”を与えると18が返る。  
関数convDecimalが利用する関数intは，引数で与えられた文字が“0”なら整数値0を返し，“1”なら整数値1を返す。

### 〔プログラム〕
<pre>

○整数型: convDecimal(文字列型: binary)
　整数型: i, length, result ← 0
　length ← binaryの文字数
　for (iを 1 から lengthまで 1 ずつ増やす)
　  result ← <span style="border: solid 1px">　　　</span>
　endfor
　return result
</pre>

### 解答群
ア result ＋ int(binaryの(length－i＋1)文字目の文字)  
イ result ＋ int(binaryのi文字目の文字)  
ウ result × 2 ＋ int(binaryの(length－i＋1)文字目の文字)  
エ result × 2 ＋ int(binaryのi文字目の文字)

```python
def convDecimal(binary):
    result = 0
    length = len(binary)
    for i in range(1, length + 1):
        result = result * 2 + int(binary[i - 1])
    return result
```
