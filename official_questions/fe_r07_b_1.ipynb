### 問1 同じ値を返す関数
次のプログラム中の <span style="border: solid 1px">　a　</span> と <span style="border: solid 1px">　b　</span> に入れる正しい答えの組合せを，解答群の中から選べ。

関数function1が受け取る引数と，関数function2が受け取る引数とが同じとき，二つの関数は同じ値を返す。ここで，引数nと引数mは正の整数であり，引数mは引数nよりも10以上大きい数とする。
### 〔プログラム〕
<pre> 
○整数型: function1(整数型: n, 整数型: m)
　整数型: count ← 0
　整数型:i 
　for (i を n から m まで 1 ずつ増やす) 
　　if ((i mod 4) が 0 と等しい) 
　　　count ← count＋ 1 
　　endif 
　endfor
　return count 

○整数型: function2(整数型: n, 整数型: m)
　整数型: count ← 0
　整数型: tempN ← n
　整数型: i, j 
　for ( <span style="border: solid 1px">　a　</span> ) 
　　if ((tempN mod 4) が 0 と等しい)
　　　繰返し処理を終了する 
　　endif 
　　tempN ← tempN＋ 1 
　endfor 

　for ( <span style="border: solid 1px">　b　</span> ) 
　　count ← count＋ 1 
　endfor 
　return count
</pre>

### 解答群
<table><tr><th></th><th>a</th><th>b</th></tr>
<tr><td>ア</td><td>iを1から2まで1ずつ増やす</td><td>jをnから始めてmを超えない範囲で tempNずつ増やす</td></tr>
<tr><td>イ</td><td>iを1から2まで1ずつ増やす</td><td>jをtempNからmまで1ずつ増やす</td></tr>
<tr><td>ウ</td><td>iを1から2まで1ずつ増やす</td><td>jをtempNから始めてmを超えない範囲で4ずつ増やす</td></tr>
<tr><td>エ</td><td>iを1から3まで1ずつ増やす</td><td>jをnから始めてmを超えない範囲で tempNずつ増やす</td></tr>
<tr><td>オ</td><td>iを1から3まで1ずつ増やす</td><td>jをtempNからmまで1ずつ増やす</td></tr>
<tr><td>カ</td><td>iを1から3まで1ずつ増やす</td><td>jをtempNから始めて mを超えない範囲で4ずつ増やす</td></tr></table>

```python
def function1(n, m):
    count = 0
    i = None
    for i in range(n, m + 1):
        if i % 4 == 0:
            count = count + 1
    return count
```

```python
def function2(n, m):
    count = 0
    tempN = n
    i = None
    j = None
    for i in range(1, 3 + 1):
        if tempN % 4 == 0:
            break
        tempN = tempN + 1
    for j in range(tempN, m + 1, 4):
        count = count + 1
    return count
```

```python
# 引数nとmを定義
n = 10
m = 20

# function1とfunction2を呼び出して結果を表示
result1 = function1(n, m)
result2 = function2(n, m)
print("function1の結果:", result1)
print("function2の結果:", result2)
```
