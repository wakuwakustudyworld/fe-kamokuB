### 問13
次の記述中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。ここで，配列の要素番号は1から始まる。

　関数searchは，引数dataで指定された配列に，引数targetで指定された値が含まれていればその要素番号を返し，含まれていなければ －1 を返す。dataは昇順に整列されており，値に重複はない。  
　関数search には不具合がある。例えば，data の <span style="border: solid 1px">　　　</span> 場合は，無限ループになる。

### 〔プログラム〕  
<pre>
○整数型: search(整数型の配列: data, 整数型: target)
　整数型: low, high, middle
　low ← 1
　high ← dataの要素数
　while (low ≦ high)
　 middle ← (low ＋ high) ÷ 2 の商
　 if (data[middle] ＜ target)
　 low ← middle
　 elseif (data[middle] ＞ target)
　 high ← middle
　 else
　 return middle
　 endif
　endwhile
　return －1
</pre>

### 解答群
ア 要素数が1 で，target がその要素の値と等しい  
イ 要素数が2 で，target がdata の先頭要素の値と等しい  
ウ 要素数が2 で，target がdata の末尾要素の値と等しい  
エ 要素に－1 が含まれている

```python
# 配列の要素番号を1から始めるためのクラス
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def search(data : Array, target : int) -> int:
    low = 1
    high = len(data)
    while (low <= high):
        middle = (low + high) // 2
        if (data[middle] < target):
            low = middle
        elif (data[middle] > target):
            high = middle
        else:
            return middle
    return -1
```

### 解答群実行テスト  
ア　要素数が1で，targetがその要素の値と等しい

```python
search(Array.from_literal([1]), 1) 
```

イ　要素数が2で，targetがdataの先頭要素の値と等しい

```python
search(Array.from_literal([1, 2]), 1) 
```

ウ　要素数が2で，targetがdataの末尾要素の値と等しい  
注意：無限ループになるので，割り込みで中断する必要あり

```python
search(Array.from_literal([1, 2]), 2) 
```

エ　要素に-1が含まれている

```python
search(Array.from_literal([-1, 2]), -1) 
```
