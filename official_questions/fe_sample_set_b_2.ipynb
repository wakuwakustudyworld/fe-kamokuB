## 基本情報技術者試験（科目B試験）サンプル問題セット
### 問2
次のプログラム中の <span style="border: solid 1px">　a　</span> 〜 <span style="border: solid 1px">　c　</span> に入れる正しい答えの組合せを，解答群の中から選べ。

関数fizzBuzzは，引数で与えられた値が，3 で割り切れて 5 で割り切れない場合は “3 で割り切れる” を，5 で割り切れて 3 で割り切れない場合は “5 で割り切れる” を，3 と 5で割り切れる場合は “3 と 5で割り切れる” を返す。それ以外の場合は “3 でも 5でも割り切れない” を返す。

### 〔プログラム〕
<pre>
○文字列型: fizzBuzz(整数型: num)
　文字列型: result 
　if (numが <span style="border: solid 1px">　a　</span> で割り切れる )  
　  result ←  "<span style="border: solid 1px">　a　</span> で割り切れる"  
　elseif (numが <span style="border: solid 1px">　b　</span> で割り切れる)  
　  result ←  "<span style="border: solid 1px">　b　</span> で割り切れる"  
　elseif (numが <span style="border: solid 1px">　c　</span> で割り切れる)  
　  result ←  "<span style="border: solid 1px">　c　</span> で割り切れる"  
　else  
　  result ← "3 でも 5 でも割り切れない"
　endif
　return result
</pre>

### 解答群

||a|b|c|
|---|---|---|---|
|ア|3|3 と 5|5|
|イ|3|5|3 と 5|
|ウ|3 と 5|3|5|
|エ|5|3|3と 5|
|オ|5|3 と 5|3|  

```python
def fizzBuzz(num : int) -> str:
    if num % (3*5) == 0:  # 数字の部分が空欄a
        result = "3と5で割り切れる" # 数字の部分が空欄a
    elif num % 3 == 0: # 数字の部分が空欄b
        result = "3で割り切れる" # 数字の部分が空欄b
    elif num % 5 == 0: # 数字の部分が空欄c
        result = "5で割り切れる" # 数字の部分が空欄c
    else:
        result = "3でも5でも割り切れない"
    return result
```

```python
print(fizzBuzz(15))
```
