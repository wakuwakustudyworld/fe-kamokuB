### 問6
次のプログラム中の <span style="border: solid 1px">　　　</span> に入れる正しい答えを，解答群の中から選べ。

　関数revは8ビット型の引数byteを受け取り，ビットの並びを逆にした値を返す。  
　例えば，関数revをrev(01001011)として呼び出すと，戻り値は 11010010 となる。
なお，演算子 ∧ はビット単位の論理積，演算子 ∨ はビット単位の論理和，演算子 >> は論理右シフト，演算子 << は論理左シフトを表す。例えば，value >> n はvalue の値をnビットだけ右に論理シフトし，value << n はvalue の値をnビットだけ左に論理シフトする。

### 〔プログラム〕
<pre>
○8ビット型: rev(8 ビット型: byte)
　8ビット型: rbyte ← byte
　8ビット型: r ← 00000000
　整数型: i
　for (iを1から8まで1ずつ増やす)
    <span style="border: solid 1px">　　　</span>
　endfor
　return r
</pre>

### 解答群
ア r ← (r << 1) ∨ (rbyte ∧ 00000001)  
　　rbyte ← rbyte >> 1  
イ r ← (r << 7) ∨ (rbyte ∧ 00000001)  
　　rbyte ← rbyte >> 7  
ウ r ← (rbyte << 1) ∨ (rbyte >> 7)  
　　rbyte ← r  
エ r ← (rbyte >> 1) ∨ (rbyte << 7)  
　　rbyte ← r

※Pythonでは8ビット型はなく，「0b」で2進数を表現，bin()関数で2進数表示

```python
def rev(byte):
    rbyte = byte
    r = 0b00000000
    for i in range(1, 9):
        r = (r << 1) | (rbyte & 0b00000001) # for文内の2行が選択肢
        rbyte = rbyte >> 1
    return r
```

選択肢（Pythonで変換）  
ア　r = (r << 1) | (rbyte & 0b00000001)  
　　rbyte = rbyte >> 1  
イ　r = (r << 7) | (rbyte & 0b00000001)  
　　rbyte = rbyte >> 7  
ウ　r = (rbyte << 1) | (rbyte >> 7)  
　　rbyte = r  
エ　r = (rbyte >> 1) | (rbyte << 7)  
　　rbyte = r

```python
# 例
print(bin(rev(0b00000101)))
```
