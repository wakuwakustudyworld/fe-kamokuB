### 問1 素数を求める関数
　次のプログラム中の <span style="border: solid 1px">　a　</span> と <span style="border: solid 1px">　b　</span> に入れる正しい答えの組合せを，解答群の中から選べ。ここで，配列の要素番号は1から始まる。

　関数findPrimeNumbersは，引数で与えられた整数以下の，全ての素数だけを格納した配列を返す関数である。ここで，引数に与える整数は2以上である。

### 〔プログラム〕 
<pre>
○整数型の配列: findPrimeNumbers(整数型: maxNum) 
　整数型の配列: pnList ← {}  // 要素数0の配列 
　整数型: i, j 
　論理型: divideFlag  
　for (i を 2 から <span style="border: solid 1px">　a　</span> まで 1 ずつ増やす)  
　  divideFlag ← true  

　  /* iの正の平方根の整数部分が 2未満のときは，繰返し処理を実行しない */ 
　  for (j を 2 から iの正の平方根の整数部分 まで 1 ずつ増やす) // α 
　    if ( <span style="border: solid 1px">　b　</span> ) 
　      divideFlag ← false
　      αの行から始まる繰返し処理を終了する
　    endif
　  endfor
　  if (divideFlagが trueと等しい) 
　    pnListの末尾に iの値を追加する 
　  endif
　endfor
　return pnList
</pre>

### 解答群
<table border="1"><tr><th></th><th>a  </th><th>b</th></tr> 
<tr><td>ア</td><td>maxNum</td><td>i ÷ j の余り が 0 と等しい</td></tr> 
<tr><td>イ</td><td>maxNum</td><td>i ÷ j の商 が 1 と等しくない</td></tr> 
<tr><td>ウ</td><td>maxNum ＋ 1</td><td>i ÷ j の余り が 0 と等しい </td></tr>
<tr><td>エ</td><td>maxNum ＋ 1</td><td>i ÷ j の商 が 1 と等しくない</td></tr></table>  

```python
# 配列を1-basedで扱うためのArrayクラスをインポート
import sys
sys.path.append('..')

from one_based_array import Array
```

```python
def findPrimeNumbers(maxNum):
    pnList = Array.from_literal([])
    i = None
    j = None
    divideFlag = None
    for i in range(2, maxNum + 1):
        divideFlag = True
        for j in range(2, int(i ** 0.5) + 1):
            if i % j == 0:
                divideFlag = False
                break
        if divideFlag == True:
            pnList.append(i)
    return pnList
```

```python
# プログラムの実行例
maxNum = 20
primeNumbers = findPrimeNumbers(maxNum)
print(f"{maxNum}以下の素数: {primeNumbers}")
```
