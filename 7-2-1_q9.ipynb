## 7-2-1　予想問題
### 問9　単方向リストへの昇順挿入
#### 表 クラスListElementのメンバ変数の説明

|メンバ変数|型|説明|
|---|---|---|
|val|文字型|要素の値|
|next|ListElement|次の要素の参照<br>次の要素がないときの状態は未定義|

### 〔プログラム〕
<pre>
○ListElement: insertToSortedList(ListElement: head, ListElement: newNode)
　ListElement: current, previous
　
　if (head = 未定義 または head.val > newNode.val)
　　newNode.next ← head
　　return newNode
　endif
　
　current ← head
　previous ← 未定義
　
　while (current ≠ 未定義 かつ <span style="border: solid 1px">　a　</span>)
　　previous ← current
　　current ← current.next
　endwhile
　
　<span style="border: solid 1px">　b　</span> ← newNode
　newNode.next ← current
　
　return head
</pre>

クラスElementの定義

```python
class ListElement:
    def __init__(self, qVal):
        self.val = qVal
        self.next = None
```

```python
def insertToSortedList(head: ListElement, newNode: ListElement) -> ListElement:
    if head == None or head.val > newNode.val:
        newNode.next = head
        return newNode
    current = head
    previous = None
    while current != None and current.val < newNode.val:
        previous = current
        current = current.next
    previous.next = newNode
    newNode.next = current
    return head
```

```python
# 実行例
head = ListElement(10)
head = insertToSortedList(head, ListElement(5))
head = insertToSortedList(head, ListElement(15))
```

```python
# リストを順に出力
current = head
while current != None:
    print(current.val, end=' ')
    current = current.next
```
